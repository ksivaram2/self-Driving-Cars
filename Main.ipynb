```python
import matplotlib.pyplot as plt
```

```python
# Define the car class
class Car:
    def __init__(self):
        self.position = 0

    def drive(self, speed):
        self.position += speed
# Create a car instance
car = Car()

# Simulate the car driving
positions = []
speeds = [2, 3, 4, 5]  # Sample speeds

for speed in speeds:
    car.drive(speed)
    positions.append(car.position)
# Plot the car's positions
plt.plot(positions)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Self-Driving Car Simulation")
plt.show()
```

```python
import time

class Car:
  def __init__(self):
    self.speed=0
    self.steering_angle=0
  def set_speed(self,speed):
    self.speed=speed
  def set_steering_angle(self,angle):
    self.steering_angle=angle
  def update(self):
      print(f"Car speed: {self.speed}mph, Steering angle:{self.steering_angle} degrees")
      time.sleep(1)
car=Car()

car.set_speed(30)
car.set_steering_angle(15)
car.update()
```
